```python
import numpy as np
import pandas as pd
import xarray as xr
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
#define filepath for dataset
axps_url ='http://erddap.dataexplorer.oceanobservatories.org/erddap/tabledap/ooi-rs03axps-sf03a-2a-dofsta302.csv?time%2Cz%2Cmoles_of_oxygen_per_unit_mass_in_sea_water_profiler_depth_enabled&time%3E=2019-05-03T02%3A12%3A17Z&time%3C=2022-06-21T10%3A02%3A04Z&z%3E=-193&z%3C=91'
#renname columns to correct formatting problems
columns= ["Time", "Depth", "Dissolved O2"]
#time in UTC, depth in m, dissolved o2 in micromoles per kg

#Read file with dates as the index in datetime. Skipped colum labels and row below column labels with units and assigned new column names
axps_odata = pd.read_csv(axps_url, skiprows=2, names=columns, index_col=0, parse_dates=True)
```

```python
#check axps_odata
display(axps_odata)
```

```python
#check to see depth range and number of depths
np.unique(axps_odata['Depth'])
```

```python
#make new columns to separate out hour, day, month, and year in order to index for a given day/period of time without seconds and nanoseconds
axps_odata["Day"]=axps_odata.index.day.values
axps_odata["Year"]=axps_odata.index.year.values
axps_odata["Month"]=axps_odata.index.month.values
axps_odata["Hour"]= axps_odata.index.hour.values
```

```python
#check new columns for year, month, day, hour
display(axps_odata)
```

```python
#use year, month, day colums to create data subsets for a specific day
df_subset=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==6) & (axps_odata['Day']==25)]
df_subset_dec=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==12) & (axps_odata['Day']==25)]
```

```python
#make plot to get a better idea of data for each day
#used this code/graph to check each new daily subset- some did not have usual 9 daily profiles, days without usual 9 profiles were not used


x=df_subset.index.values
y1= df_subset['Depth']
y2= df_subset['Dissolved O2']

fig,axs =plt.subplots(figsize=(10,6))

#plot depth
axs.plot(x, y1, label='Depth')

#axs.plot(x, y2, marker=".", color='red')
axs.set_xlabel('Time')
axs.set_ylabel('Depth')
twinax= axs.twinx()
twinax.plot(x, y2, marker='.', color= 'red', label='DO2')
twinax.set_ylabel("Do2")
fig.legend()
```

```python
#average dissolved oxygen values for 1 day at a given depth to get 1 depth profile
groupby_depth=df_subset.groupby('Depth').mean(numeric_only=True)
groupby_depth_dec=df_subset_dec.groupby('Depth').mean(numeric_only=True)

#check new averaged profile
display(groupby_depth)
```

```python
#make plot of average O2 profile for 6/25/19
fig,axs =plt.subplots(figsize=(10,8))
x= groupby_depth['Dissolved O2']
y= groupby_depth.index
axs.plot(x,y, color="lightseagreen", linewidth='3', label='6/25/2019')

#make plot of average O2 profile for 12/25/19
x2= groupby_depth_dec['Dissolved O2']
y2= groupby_depth_dec.index
axs.plot(x2, y2, color='k', linewidth='3', label= '12/25/2019')



#add and format title, axis titles, grid, tickmarks, and legend
axs.set_xlabel('Dissolved Oxygen (micromoles/kg)', fontsize=15, fontweight='bold')
axs.set_ylabel('Depth (m)', fontsize=15, fontweight='bold')
axs.set_title('Daily Average Dissolved Oxygen Profile at AXPS', fontsize=20, fontweight='bold')
axs.grid()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)

#save figure
plt.savefig("Final Project Fig. 1.jpg")
```

```python
#making graph of depth vs time of profiler shuttle, simmilar to graph I used to check profiles for each day

#import package to be able to modify x axis lables (which are in datemtime)
import matplotlib.dates as mdates

#creating figure
fig,axs =plt.subplots(figsize=(10,6))
#adding titles and axis titles
axs.set_title('Depth of AXPS Shuttle Over Time (6/25/2019)', fontsize=20, fontweight='bold')
axs.set_xlabel('Time (hours)', fontsize=17, fontweight='bold')
axs.set_ylabel('Depth (meters)', fontsize=17, fontweight='bold')
#plotting depth vs. time 
axs.plot(day_subset_time, subset_depth, color="teal", linewidth='3')
#adding grid, formatting tick label size
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
axs.grid()

#rewriting x tick marks to declutter x axis and only show hours
axs.xaxis.set_major_formatter(mdates.DateFormatter("%H"))

#saving figure
plt.savefig("Depth vs. Time.jpg")


#other strategies I tried to change x axis to hours only that didn't work

#defining new x variable to be only hours instead of showing repeated day, month, year etc. for a single day
#hours_only= df_subset['Hour']
#print(hours_only)
                
# x= np.unique(hours_only)
# print(x)
# #x= ["0:00" '2:00', '4:00','6:00','8:00', '10:00', '12:00', '14:00', '16:00','18:00','20:00','22:00','24:00',]
# a=[0,1,2,3,]
# #axs.set_xticks(a)
# axs.set_xticklabels(x)
```

```python
#making graph of O2 profile day vs night

#sunrise/sunset times in PST
#December 25, 2019: sunrise at 7:54 AM (round to 8:00 AM), sunset at 4:24 PM (round to 4:00 PM) PST
#June 25, 2019: Sunrise at 5:10 (round to 5:00 AM), sunsest at 9:13 PM (round to 9:00 PM) PST

#Sunrise/sunset times converted to UTC (data in UTC)
#December 25: sunrise at 16:00 UTC, sunset at 00:00 UTC
#June 25, 2019: sunrise at 12:00 UTC, sunset at 4:00 UTC

#create new subsets for day and night for the 25th of June and December
#only worked when subsetting directly from original axps_odata, didn't work when I tried to subset from df_subset

jun_daytime=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==6) & (axps_odata['Day']==25) & ( (axps_odata['Hour']>=12) | (axps_odata['Hour']<=4))]

jun_nighttime=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==6) & (axps_odata['Day']==25) & (axps_odata['Hour']>=4) & (axps_odata['Hour']<=12)]

dec_daytime=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==12) & (axps_odata['Day']==25) & (axps_odata['Hour']>=16)]

dec_nighttime=axps_odata.loc[(axps_odata['Year']==2019) & (axps_odata['Month']==12) & (axps_odata['Day']==25) & (axps_odata['Hour']<=16)]

#check new subsets
display(jun_daytime)
display(jun_nighttime)
display(dec_daytime)
display(dec_nighttime)



```

```python
#average O2 across depth for night and day in December and June
groupby_dec_day=dec_daytime.groupby('Depth').mean(numeric_only=True)
groupby_dec_night=dec_nighttime.groupby('Depth').mean(numeric_only=True)
groupby_jun_day=jun_daytime.groupby('Depth').mean(numeric_only=True)
groupby_jun_night=jun_nighttime.groupby('Depth').mean(numeric_only=True)

#check averaged profiles
display(groupby_jun_night)
display(groupby_jun_day)
display(groupby_dec_night)
display(groupby_dec_day)
```

```python
#plot average profile for night and day in December

#making plot of average O2 profile for 12/25/19 day
fig,axs2 =plt.subplots(figsize=(10,8))
x3= groupby_dec_day['Dissolved O2']
y3= groupby_dec_day.index
axs2.plot(x3, y3, color='lightseagreen', linewidth='3', label= 'Day')

#making plot of average O2 profile for 12/25/19 night
x4= groupby_dec_night['Dissolved O2']
y4= groupby_dec_night.index
axs2.plot(x4, y4, color='k', linewidth='3', label= 'Night')

#adding and formating title, axis titles, grid, tickmarks, and legend

axs2.set_xlabel('Dissolved Oxygen (micromoles/kg)', fontsize=15, fontweight='bold')
axs2.set_ylabel('Depth (m)', fontsize=15, fontweight='bold')
axs2.set_title('Average Dissolved Oxygen Profile on 12/25/2019', fontsize=20, fontweight='bold')
axs2.grid()
plt.yticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)

#save figure
plt.savefig("Final Project Fig. 3.jpg")
```

```python
#plot averaged profile for day and night in June

#making plot of average O2 profile for 6/25/19 day
fig,axs1 =plt.subplots(figsize=(10,8))
x= groupby_jun_day['Dissolved O2']
y= groupby_jun_day.index
axs1.plot(x,y, color="lightseagreen", linewidth='3', label='Day')

#adding plot of average O2 profile for 6/25/19 night
x2= groupby_jun_night['Dissolved O2']
y2= groupby_jun_night.index
axs1.plot(x2,y2, color="k", linewidth='3', label='Night')

#adding and formating title, axis titles, grid, tickmarks, and legend
#plot for June
axs1.set_xlabel('Dissolved Oxygen (micromoles/kg)', fontsize=15, fontweight='bold')
axs1.set_ylabel('Depth (m)', fontsize=15, fontweight='bold')
axs1.set_title('Average Dissolved Oxygen Profile on 6/25/19', fontsize=20, fontweight='bold')
axs1.grid()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)

#save figure
plt.savefig("Final Project Fig. 4.jpg")
```
